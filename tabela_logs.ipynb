```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o CSV com os dados
df = pd.read_csv('logs_coletados.csv')

# Garantir que o timestamp seja tratado como datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# --- 1. Gráfico: Variação de Preço ao Longo do Tempo por Destino ---
plt.figure(figsize=(12, 6))
sns.lineplot(x='timestamp', y='price', hue='destination', data=df, marker='o')
plt.title('Variação de Preço ao Longo do Tempo por Destino')
plt.xlabel('Data e Hora da Pesquisa')
plt.ylabel('Preço (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 2. Gráfico: Distribuição de Preços por Destino ---
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='price', hue='destination', kde=True, bins=30)
plt.title('Distribuição de Preços por Destino')
plt.xlabel('Preço (R$)')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()

# --- 3. Gráfico: Preço Médio por Destino ---
plt.figure(figsize=(12, 6))
price_mean = df.groupby('destination')['price'].mean().reset_index()
sns.barplot(x='destination', y='price', data=price_mean, palette='viridis')
plt.title('Preço Médio por Destino')
plt.xlabel('Destino')
plt.ylabel('Preço Médio (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Carregar o CSV com os dados
df = pd.read_csv('logs_coletados.csv')

# Garantir que o timestamp seja tratado como datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Função para extrair o destino da URL
def extrair_destino(url):
    """
    Função que extrai o destino da URL, por exemplo, 'FLN-MAD', 'FLN-LIS', etc.
    """
    parts = url.split('/')
    return parts[4] if len(parts) > 4 else "desconhecido"

# Função para extrair a data da viagem da URL
def extrair_data_viagem(url):
    """
    Função que extrai a data da viagem da URL, por exemplo, '2025-04-19'.
    """
    parts = url.split('/')
    return parts[5] if len(parts) > 5 else "desconhecida"

# Função para obter o menor preço por destino
def obter_menor_preco_por_destino(df):
    destinos_unicos = df['destination'].unique()  # Destinos únicos
    resultados = []
    
    for destino in destinos_unicos:
        # Filtrar os dados do DataFrame para o destino específico
        destino_df = df[df['destination'] == destino]
        
        if destino_df.empty:
            print(f"Não foram encontrados dados para o destino {destino}.")
            continue

        # Encontrar o menor preço para esse destino
        menor_preco = destino_df['price'].min()
        
        # Encontrar a URL correspondente ao menor preço
        url_menor_preco = destino_df.loc[destino_df['price'] == menor_preco, 'url'].iloc[0]
        
        # Extrair a data da viagem da URL
        data_viagem = extrair_data_viagem(url_menor_preco)
        
        # Adicionar o resultado na lista
        resultados.append({
            'Destino': destino,
            'Data da Viagem': data_viagem,
            'Preço': menor_preco
        })
    
    return resultados

# Obter os resultados
resultados = obter_menor_preco_por_destino(df)

# Exibir os resultados
for resultado in resultados:
    print(f"Destino: {resultado['Destino']}, Data da Viagem: {resultado['Data da Viagem']}, Preço: R$ {resultado['Preço']}")
```

```python
import pandas as pd

# Caminhos dos arquivos CSV
csv_file_1 = "logs_coletados_1.csv"
csv_file_2 = "logs_coletados_2.csv"

# Carregar os dois CSVs em DataFrames separados
df_1 = pd.read_csv(csv_file_1)
df_2 = pd.read_csv(csv_file_2)

# Concatenar os dois DataFrames
df = pd.concat([df_1, df_2], ignore_index=True)

# Garantir que o timestamp seja tratado como datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Função para extrair o destino da URL
def extrair_destino(url):
    """
    Função que extrai o destino da URL, por exemplo, 'FLN-MAD', 'FLN-LIS', etc.
    """
    parts = url.split('/')
    return parts[4] if len(parts) > 4 else "desconhecido"

# Função para extrair a data da viagem da URL
def extrair_data_viagem(url):
    """
    Função que extrai a data da viagem da URL, por exemplo, '2025-04-19'.
    Remove qualquer informação adicional após a data (ex: ?sort=price_a).
    """
    parts = url.split('/')
    data_viagem = parts[5] if len(parts) > 5 else "desconhecida"
    # Remover qualquer parte após a data (por exemplo, ?sort=price_a)
    data_viagem = data_viagem.split('?')[0]
    return data_viagem

# Função para obter o menor preço por destino e somar ida e volta
def obter_menor_preco_por_destino(df):
    destinos_ida = ["FLN-MAD", "FLN-AMS", "FLN-LIS", "FLN-OPO", "FLN-ROM"]
    destinos_volta = ["MAD-FLN", "AMS-FLN", "LIS-FLN", "OPO-FLN", "ROM-FLN"]
    
    resultados = []
    
    for ida, volta in zip(destinos_ida, destinos_volta):
        # Filtra os dados do DataFrame para ida
        ida_df = df[df['destination'] == ida]
        if ida_df.empty:
            print(f"Não foram encontrados dados para o destino ida {ida}.")
            continue
        
        # Encontrar o menor preço para a ida
        menor_preco_ida = ida_df['price'].min()
        
        # Encontrar a URL correspondente ao menor preço de ida
        url_menor_preco_ida = ida_df.loc[ida_df['price'] == menor_preco_ida, 'url'].iloc[0]
        data_viagem_ida = extrair_data_viagem(url_menor_preco_ida)

        # Filtra os dados do DataFrame para volta
        volta_df = df[df['destination'] == volta]
        if volta_df.empty:
            print(f"Não foram encontrados dados para o destino volta {volta}.")
            continue
        
        # Encontrar o menor preço para a volta
        menor_preco_volta = volta_df['price'].min()
        
        # Encontrar a URL correspondente ao menor preço de volta
        url_menor_preco_volta = volta_df.loc[volta_df['price'] == menor_preco_volta, 'url'].iloc[0]
        data_viagem_volta = extrair_data_viagem(url_menor_preco_volta)
        
        # Adicionar o resultado
        soma_preco = menor_preco_ida + menor_preco_volta
        resultados.append({
            'Destino': f"{ida} - {volta}",
            'Menor Preço Ida': menor_preco_ida,
            'Data Ida': data_viagem_ida,
            'Menor Preço Volta': menor_preco_volta,
            'Data Volta': data_viagem_volta,
            'Soma Preços': soma_preco,
            'Mensagem': f"O valor total para ir no dia {data_viagem_ida} e voltar no dia {data_viagem_volta} para {ida}, é de R$ {soma_preco}"
        })
    
    return resultados

# Obter os resultados
resultados = obter_menor_preco_por_destino(df)

# Exibir os resultados
for resultado in resultados:
    print(f"Destino: {resultado['Destino']}")
    print(f"Menor Preço Ida: R$ {resultado['Menor Preço Ida']} - Data Ida: {resultado['Data Ida']}")
    print(f"Menor Preço Volta: R$ {resultado['Menor Preço Volta']} - Data Volta: {resultado['Data Volta']}")
    print(f"Soma dos Preços: {resultado['Mensagem']}")
    print("-" * 50)
```
